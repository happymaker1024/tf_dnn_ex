# 피마인디언의 당뇨병 예측(DNN 2진분류)

```python
import pandas as pd
```

```python
data_path = "datas_dnn/pima-indians-diabetes3.csv"
df = pd.read_csv(data_path)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

# X, y 데이터 준비

```python
X = df.loc[:, 'pregnant':'age']
y = df['diabetes']
```

```python
X.shape
```

```python
y.shape
```

# 모델 구조 설계

- 입력데이터(features) : 8개
- 출력(target) : 2진분류 (1개 노드)

```python
# 라이브러리 임포트
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
# 모델 구조 만들기
model = Sequential()
model.add(Input(shape=(8,)))
model.add(Dense(12, activation='relu', name="Dense_1"))
model.add(Dense(8, activation='relu', name="Dense_2"))
model.add(Dense(1, activation='sigmoid', name="Dense_3"))
```

```python
model.summary()
```

# 학습 방식 정의

```python
model.compile(
    loss="binary_crossentropy", 
    optimizer='adam',
    metrics=['accuracy']
)
```

# 학습 진행

```python
model.fit(
    X, y,  # <- 학습데이터
    epochs = 100,  # 전체 데이터 학습횟수
    batch_size = 5,  # 한번 학습학때 5개 데이터 셋을 학습함. 1 iteration
    # verbose = 'auto'
)
```
